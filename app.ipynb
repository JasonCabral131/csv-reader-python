```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the updated CSV file with semicolon as the delimiter
df = pd.read_csv('sales.csv', sep=';')

# Plotting with names
plt.plot(df['Sales'], marker='o')
plt.xlabel('Person Index')
plt.ylabel('Sales')
plt.title('Sales Plot')

for i, row in df.iterrows():
    plt.text(i, row['Sales'], f"{row['First name']} {row['Last name']}", ha='right', va='bottom')

plt.show()

best_sales_person = df.loc[df['Sales'].idxmax()]

print("Person with the best sales:")
print(best_sales_person[['First name', 'Last name', 'Sales']])
```
